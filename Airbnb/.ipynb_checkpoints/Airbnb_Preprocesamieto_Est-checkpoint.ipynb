# Etapa 1. Limpieza y preprocesamiento de datos  comunidad de Airbnb! Datos públicos no procesados

```python
# Liberrias
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Datos de Madrid España- Listado de Publicaciones de Airbnb
data= pd.read_csv('listings.csv')
```

```python
#Verifico la cantidad de datos que hay en el dataset
data.shape
```

```python
data.columns
```

```python
#Verifico los datos faltantes del dataset o nulos
data.isnull().sum()
```

```python
#Procesamiento de los datos**
#Convertir los precios en flotantes, caso del precio esta en objeto con el simbolo$
data.price.head()
```

```python
data['price'] = data['price'].str.replace( '[$, ,]', '').astype(float)
data.price.head()
```

```python
#Convertir los depositos de seguridad en flotantes, caso del precio esta en objeto con el simbolo$
data.security_deposit.head()
```

```python
data.security_deposit.dtype
```

```python
#Convertir los depositos de seguridad en flotantes y convertir los nulos en ceros.#sale error porque ya esta covertidos en float
data['security_deposit'] = data['security_deposit'].str.replace( '[$, ,]', '').astype(float)
```

```python
#Rellena las cledas nula con cero
data['security_deposit'] = data['security_deposit'].fillna(value=0)
data.security_deposit.head()
```

```python
#Convertir los cargos por limpieza en flotantes y convertir los nulos en ceros.#sale error porque ya esta covertidos en float
data['cleaning_fee'] = data['cleaning_fee'].str.replace( '[$, ,]', '').astype(float)
```

```python
data.cleaning_fee.head()
```

```python
#Rellenar los flotantes con cero
data['cleaning_fee'] = data['cleaning_fee'].fillna(value=0)
data.cleaning_fee.head()
```

```python
# Reduccion de las publicaciones con precios > 300 dolares y <=0 aplicada a la caracteristica del precio
#ahora 425 filas menos que al principio
data.shape
data = data.loc[(data.price <=300) & (data.price >=0)]
data.shape
```

```python
#Limpiar columna de comodidades esta en puro texto y se puede convertir a una matriz
#Eilimnaos las llaves al pricpicio y al final del texto
data.amenities.head()
```

```python
#Elimiar las llaves y las comillas con instruccion: replace
#data.amenities.head()
data.amenities = data.amenities.str.replace("[{}]", "").str.replace('"',"")
data.amenities.head(15)
#ya se limpiaron los datos
```

```python
data['amenities']
```

'''
Ahora convertir los textos en una matrix y indicamos que elemestos de separación es la coma
y creamos una variable (df_amenities) creamos un nuevo dataframe con 169 columnas, la misma cantidad de filas de datos originales
y una tabla llena con 0 y 1, por le momento  no se incluyen a los datos aun.
''''

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vectorizer = CountVectorizer(tokenizer=lambda x: x.split(','))#convierte los textos en una matriz
amenities = count_vectorizer.fit_transform(data['amenities'])
df_amenities = pd.DataFrame(amenities.toarray(), columns=count_vectorizer.get_feature_names_out())
df_amenities = df_amenities.drop('',1)
```

```python
# Es posible que haya registros con True o False
df_amenities.info
```

```python
#Mezclar el dataset
df_amenities = df_amenities.sample(frac=1).reset_index(drop=True)
df_amenities.head(10)
```

```python
#observamos 169 columnas
df_amenities.shape
```

```python
df_amenities.describe()
```

Siguiente es convertir todos las columnas con datos a númericos- f=0 y t=1 (false y true) para esto se crea una variable "columnas" y guardamos todas lascolumnas con este tipo de datos
creamos una funcion para c en columas reemplace la letra f con 0 y la letra (t) con 1

```python
data.host_identity_verified.head()
```

```python
# Conversión todos las columnas con datos a númericos- f=0 y t=1 (false y true)
columnas = ['host_is_superhost','host_identity_verified', 'host_has_profile_pic',
            'is_location_exact', 'requires_license', 'instant_bookable', 
            'require_guest_profile_picture', 'require_guest_phone_verification']

for c in columnas:
    data[c] = data[c].replace('f',0,regex=True)
    data[c] = data[c].replace('t',1,regex=True)

data.host_identity_verified.head()          
```

- Las anterior columnas que tenian f y t ahora se han convettiod en 0 y 1

Rellenar los datos faltantes con la Media. para ello se crea la variable "datares" en donde se guardan los valores faltantes con la media de la misma. Y hacemos una función para verificar si en las columnas definidas hay algun valor faltante. si es  asi se llena con la media de la columna

```python
datares = data[['host_is_superhost','host_identity_verified', 'host_has_profile_pic',
            'is_location_exact', 'requires_license', 'instant_bookable', 
            'require_guest_profile_picture', 'require_guest_phone_verification', 'security_deposit',
            'cleaning_fee','host_listings_count','host_total_listings_count','property_type','bathrooms','bedrooms','guests_included',
            'number_of_reviews','review_scores_rating','minimum_nights', 'price']]

for col in datares.columns[datares.isnull().any()]:
    datares[col] = datares[col].fillna(datares[col].median())
```

- en lo anterior, Se utilizo isnull para ver si la columna tiene valores faltantes y fillna para llenar la c eldas con valor indicado en este caso con la media y lo datos de cargab enla variable datares que almacena parat de los datos de la data original

```python
data.isnull().sum()
```

```python
#revisar como queda datares
datares.shape
```

```python
df_amenities.shape
```

Vamos unir los  datos finales, estan en datares con la columan de comodidades(df_amenities), instrucción "concat" ahora el dataset cuneta con 189 columnas

```python
datafin = pd.concat([datares, df_amenities], axis=1, join="inner")
datafin.shape
```

```python
#Ahora se cuenta con 189 columnas el nuevo dataset, se ha eliminado filas que no compaerte el mimso argumento
#Ya se tiene un dataset bastnate limpio
datafin
```

```python
#Seguir trabajando en las demas columnas para mejorar resualtdos
datafin.review_scores_rating.head()
```

```python
#Guardamos el dataset- Hacemos una clasificacion binaria.
datafin.to_csv('datares_pro.csv', index=False)
```

```python
#Ahora eempezamos a construir el modelo, separamos los datos de entrenamiento y pruebas
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state=42)
```

```python
#Entrenar el modelos juntos con los datos de entrenamiento y el método
RF.fit(X_train, y_train)
```

```python
#realizar una prediccion para ver que tal es el modelo que se construyo
y_train_pred = RF.predict(X_train)
y_test_pred = RF.predict(X_test)
```

```python
'''# Calculo de la Predicción del Modelo
'''
# aplicando metrica de Error promedio al cuadrado. Esta metrica es bueno si el Numro esta cercano al 0
rmse_rf = (mean_squared_error(y_test,y_test_pred))**(1/2)
print('Error promedio al cuadrado :', rmse_rf)

#calculo R2 #un buen resulatdo es mas cercano al uno mucho mejor
print('R2 :', r2_score(y_test, y_test_pred))

#*** Se obsrev que el modelo es regular***
```
